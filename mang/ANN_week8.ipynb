## **Mang ANN voi Iris**

```python
from sklearn import datasets
from sklearn.linear_model import Perceptron 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target 
print(X[:10])
print(y[:10])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
# scale du lieu

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train) # scale X_train
X_test_std = sc.transform(X_test)   # scale X_test
```

**hoc khong Scale**

```python
model = Perceptron(max_iter = 1000, eta0 = 0.1, random_state = 0)
# (so lan hoc, toc do hoc, ngau nhien=0)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(y_pred)
print(y_test)
print("Do chinh xac mo hinh chua scale : ", accuracy_score(y_test, y_pred))
```

```python
model = Perceptron(max_iter = 1000, eta0 = 0.1, random_state = 0)
# (so lan hoc, toc do hoc, ngau nhien=0)
model.fit(X_train_std, y_train)
y_pred = model.predict(X_test_std)
print(y_pred)
print(y_test)
print("Do chinh xac mo hinh sau scale : ", accuracy_score(y_test, y_pred))
```

# **bai 3 - lung cancer example**

```python
import pandas as pd
url_lcancer = '/content/drive/MyDrive/A_HOC TAP/Nam 4, Hoc ky 2/Artificial Intelligence/Code_AI/files/lung_cancer_examples.csv'
lcancer = pd.read_csv(url_lcancer)
lcancer.head(10)
```

```python
lcancer = lcancer.drop('Surname', axis=1)
lcancer = lcancer.drop('Name', axis=1)
lcancer.head(5)
```

```python
X = lcancer.iloc[:,:4].head(10)
y = lcancer.iloc[:,4].head(10)
X
```

```python
from sklearn import datasets
from sklearn.linear_model import Perceptron 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25)

# scale du lieu

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train) # scale X_train
X_test_std = sc.transform(X_test)   # scale X_test

model = Perceptron(max_iter = 600, eta0 = 0.6, random_state = 0)
# (so lan hoc, toc do hoc, ngau nhien=0)
model.fit(X_train_std, y_train)
y_pred = model.predict(X_test_std)
print(y_pred)
print(y_test)
print("Do chinh xac mo hinh sau scale : ", accuracy_score(y_test, y_pred))
```
