## **Fuzzy Logic - Nutritional Needs Recommendation**

```python
!pip install scikit-fuzzy

import numpy as np 
import skfuzzy as fuzzy 
from skfuzzy import control as ctrl
```

```python
height = ctrl.Antecedent(np.arange(1.0,2.1,0.1),'Height')
weight = ctrl.Antecedent(np.arange(30,91,1),'Weight')
sex = ctrl.Antecedent(np.arange(0,11,1),'Sex')
age = ctrl.Antecedent(np.arange(12,75,1),'Age')
rec = ctrl.Consequent(np.arange(0,11,1),'Recommendation')

height['L'] = fuzzy.trimf(height.universe,[1.0,1.2,1.5])
height['N'] = fuzzy.trimf(height.universe,[1.2,1.5,1.8])
height['H'] = fuzzy.trimf(height.universe,[1.5,1.8,2.0])

weight['L'] = fuzzy.trimf(weight.universe,[30,30,60])
weight['N'] = fuzzy.trimf(weight.universe,[30,60,90])
weight['H'] = fuzzy.trimf(weight.universe,[60,90,90])

sex['M'] = fuzzy.trimf(sex.universe,[0,0,7])
sex['W'] = fuzzy.trimf(sex.universe,[3,10,10])

age['L'] = fuzzy.trimf(age.universe,[12,12,40])
age['N'] = fuzzy.trimf(age.universe,[20,40,60])
age['H'] = fuzzy.trimf(age.universe,[40,75,75])

rec['L'] = fuzzy.trimf(rec.universe,[0,2,5])
rec['N'] = fuzzy.trimf(rec.universe,[2,5,8])
rec['H'] = fuzzy.trimf(rec.universe,[5,8,10])
```

```python
rule1 = ctrl.Rule(height['L']&weight['L']&sex['W']&age['L'],rec['L'])
rule2 = ctrl.Rule(height['L']&weight['L']&sex['W']&age['N'],rec['L'])
rule3 = ctrl.Rule(height['L']&weight['L']&sex['W']&age['H'],rec['L'])

rule4 = ctrl.Rule(height['L']&weight['L']&sex['M']&age['L'],rec['L'])
rule5 = ctrl.Rule(height['L']&weight['L']&sex['M']&age['N'],rec['L'])
rule6 = ctrl.Rule(height['L']&weight['L']&sex['M']&age['H'],rec['L'])

rule7 = ctrl.Rule(height['L']&weight['N']&sex['W']&age['L'],rec['L'])
rule8 = ctrl.Rule(height['L']&weight['N']&sex['W']&age['N'],rec['L'])
rule9 = ctrl.Rule(height['L']&weight['N']&sex['W']&age['H'],rec['L'])

rule10 = ctrl.Rule(height['L']&weight['N']&sex['M']&age['L'],rec['L'])
rule11 = ctrl.Rule(height['L']&weight['N']&sex['M']&age['N'],rec['L'])
rule12 = ctrl.Rule(height['L']&weight['N']&sex['M']&age['H'],rec['L'])

rule13 = ctrl.Rule(height['L']&weight['H']&sex['W']&age['L'],rec['L'])
rule14 = ctrl.Rule(height['L']&weight['H']&sex['W']&age['N'],rec['L'])
rule15 = ctrl.Rule(height['L']&weight['H']&sex['W']&age['H'],rec['L'])

rule16 = ctrl.Rule(height['L']&weight['H']&sex['M']&age['L'],rec['N'])
rule17 = ctrl.Rule(height['L']&weight['H']&sex['M']&age['N'],rec['N'])
rule18 = ctrl.Rule(height['L']&weight['H']&sex['M']&age['H'],rec['N'])
###############

rule19 = ctrl.Rule(height['N']&weight['L']&sex['W']&age['L'],rec['N'])
rule20 = ctrl.Rule(height['N']&weight['L']&sex['W']&age['N'],rec['N'])
rule21 = ctrl.Rule(height['N']&weight['L']&sex['W']&age['H'],rec['N'])

rule22 = ctrl.Rule(height['N']&weight['L']&sex['M']&age['L'],rec['N'])
rule23 = ctrl.Rule(height['N']&weight['L']&sex['M']&age['N'],rec['N'])
rule24 = ctrl.Rule(height['N']&weight['L']&sex['M']&age['H'],rec['N'])

rule25 = ctrl.Rule(height['N']&weight['N']&sex['W']&age['L'],rec['N'])
rule26 = ctrl.Rule(height['L']&weight['N']&sex['W']&age['N'],rec['N'])
rule27 = ctrl.Rule(height['N']&weight['N']&sex['W']&age['H'],rec['N'])

rule28 = ctrl.Rule(height['N']&weight['N']&sex['M']&age['L'],rec['N'])
rule29 = ctrl.Rule(height['N']&weight['N']&sex['M']&age['N'],rec['N'])
rule30 = ctrl.Rule(height['N']&weight['N']&sex['M']&age['H'],rec['N'])

rule31 = ctrl.Rule(height['N']&weight['H']&sex['W']&age['L'],rec['N'])
rule32 = ctrl.Rule(height['N']&weight['H']&sex['W']&age['N'],rec['N'])
rule33 = ctrl.Rule(height['N']&weight['H']&sex['W']&age['H'],rec['N'])

rule34 = ctrl.Rule(height['N']&weight['H']&sex['M']&age['L'],rec['N'])
rule35 = ctrl.Rule(height['N']&weight['H']&sex['M']&age['N'],rec['N'])
rule36 = ctrl.Rule(height['N']&weight['H']&sex['M']&age['H'],rec['N'])
###########

rule37 = ctrl.Rule(height['H']&weight['L']&sex['W']&age['L'],rec['H'])
rule38 = ctrl.Rule(height['H']&weight['L']&sex['W']&age['N'],rec['H'])
rule39 = ctrl.Rule(height['H']&weight['L']&sex['W']&age['H'],rec['H'])

rule40 = ctrl.Rule(height['H']&weight['L']&sex['M']&age['L'],rec['H'])
rule41 = ctrl.Rule(height['H']&weight['L']&sex['M']&age['N'],rec['H'])
rule42 = ctrl.Rule(height['H']&weight['L']&sex['M']&age['H'],rec['H'])

rule43 = ctrl.Rule(height['H']&weight['N']&sex['W']&age['L'],rec['H'])
rule44 = ctrl.Rule(height['H']&weight['N']&sex['W']&age['N'],rec['H'])
rule45 = ctrl.Rule(height['H']&weight['N']&sex['W']&age['H'],rec['H'])

rule46 = ctrl.Rule(height['H']&weight['N']&sex['M']&age['L'],rec['H'])
rule47 = ctrl.Rule(height['H']&weight['N']&sex['M']&age['N'],rec['H'])
rule48 = ctrl.Rule(height['H']&weight['N']&sex['M']&age['H'],rec['H'])

rule49 = ctrl.Rule(height['H']&weight['H']&sex['W']&age['L'],rec['H'])
rule50 = ctrl.Rule(height['H']&weight['H']&sex['W']&age['N'],rec['H'])
rule51 = ctrl.Rule(height['H']&weight['H']&sex['W']&age['H'],rec['H'])

rule52 = ctrl.Rule(height['H']&weight['H']&sex['M']&age['L'],rec['H'])
rule53 = ctrl.Rule(height['H']&weight['H']&sex['M']&age['N'],rec['H'])
rule54 = ctrl.Rule(height['H']&weight['H']&sex['M']&age['H'],rec['H'])
```

```python
height.view()
weight.view()
sex.view()
age.view()
rec.view()
```

```python
recing_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27,rule28,rule29,rule30,rule31,rule32,rule33,rule34,rule35,rule36,rule37,rule38,rule39,rule40,rule41,rule42,rule43,rule44,rule45,rule46,rule47,rule48,rule49,rule50,rule51,rule52,rule53,rule54])
recing = ctrl.ControlSystemSimulation(recing_ctrl)
recing.input['Height'] = 1.9
recing.input['Weight'] = 72
recing.input['Sex'] = 8
recing.input['Age'] = 25
recing.compute()
print(recing.output['Recommendation'])
rec.view(sim=recing)

# rainfalling = ctrl.ControlSystemSimulation(rainfalling_ctrl)
# rainfalling.input['Temp'] = 41.2
# rainfalling.input['Wind'] = 3.1
# rainfalling.compute()
# print(rainfalling.output['Rainfall'])
# rainfall.view(sim=rainfalling)
```
